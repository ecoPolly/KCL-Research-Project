```python

def dG(valid_clean):
    kT = 4.11  # pN·nm, circa 300 K
    valid_clean = np.array(valid_clean)

    # Jarzynski equality
    dG = -kT * np.log(np.mean(np.exp(-valid_clean / kT)))

    # Jackknife estimate
    jk = np.array([
        -kT * np.log(np.mean(np.exp(-np.delete(valid_clean, i) / kT)))
        for i in range(len(valid_clean))
    ])
    stima = jk.mean()
    var = (len(valid_clean) - 1) / len(valid_clean) * np.sum((jk - stima) ** 2)

    print(f"ΔG = {dG:.3f} pN·nm | Jackknife = {stima:.3f} ± {np.sqrt(var):.3f}")
    return dG

# === Inserisci qui i file che ti servono ===
files = [
    r"C:\Users\bulga\Documents\THESIS\W_dissipation\mtdata_2.csv",
    r"C:\Users\bulga\Documents\THESIS\W_dissipation\mtdata_4.csv",
    r"C:\Users\bulga\Documents\THESIS\W_dissipation\mtdata_5.csv", 
    r"C:\Users\bulga\Documents\THESIS\W_dissipation\mtdata_6.csv"
]

data = np.concatenate([
    pd.read_csv(f, usecols=[2]).dropna().values.flatten()
    for f in files
])

# === Calcola dG totale ===
dG(data)

plt.hist(data, bins= 300)
plt.xlim(40, 200)
plt.xlabel("W clean [pN nm]")
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def plot_segmented_csv(filepath):
    time = []
    x_folded = []
    x_unfolded = []

    # Leggi CSV
    with open(filepath, newline='') as f:
        reader = csv.DictReader(f)
        for row in reader:
            t = float(row["Time [s]"])
            time.append(t)

            xf = float(row["Extension_folded [nm]"]) if row["Extension_folded [nm]"] else np.nan
            xu = float(row["Extension_unfolded [nm]"]) if row["Extension_unfolded [nm]"] else np.nan

            x_folded.append(xf)
            x_unfolded.append(xu)

    time = np.array(time)
    x_folded = np.array(x_folded)
    x_unfolded = np.array(x_unfolded)

    # Plot step
    plt.figure(figsize=(8, 4))
    plt.step(time, x_folded, where="post", color="blue", alpha=0.8, linewidth=0.8, label="Folded")
    plt.step(time, x_unfolded, where="post", color="orange", alpha=0.8, linewidth=0.8, label="Unfolded")

    plt.xlabel("Time [s]")
    plt.ylabel("Extension [nm]")
    plt.title("Segmented Trajectory (Step Plot)")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()



plot_segmented_csv(r"C:\Users\bulga\Documents\THESIS\W_dissipation\TEST_PLOT_segmented.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leggi i due file Excel
file1 = 'Funf_mtdata4.csv'  # Sostituisci con il nome del tuo primo file
file2 = 'mtdata_4.csv'  # Sostituisci con il nome del tuo secondo file

# Leggi la prima colonna del primo file
df1 = pd.read_csv(file1)
x_data = df1.iloc[:, 0]  # Prima colonna (indice 0)

# Leggi la seconda colonna del secondo file
df2 = pd.read_csv(file2)
y_data = df2.iloc[:, 1]  # Seconda colonna (indice 1)

# Verifica che le lunghezze siano compatibili
min_length = min(len(x_data), len(y_data))
x_data = x_data[:min_length]
y_data = y_data[:min_length]

# Identifica gli outliers usando IQR (Interquartile Range)
def trova_outliers(data, asse='', fattore= 3):
    Q1 = data.quantile(0.20)
    Q3 = data.quantile(0.80)
    IQR = Q3 - Q1
    lower_bound = Q1 - fattore * IQR
    upper_bound = Q3 + fattore * IQR
    
    outliers_mask = (data < lower_bound) | (data > upper_bound)
    outliers_indices = data[outliers_mask].index.tolist()
    
    if outliers_indices:
        print(f"\n--- Outliers sull'asse {asse} (fattore={fattore}) ---")
        print(f"Indici: {outliers_indices}")
        print(f"Valori: {data[outliers_indices].tolist()}")
        print(f"Range normale: [{lower_bound:.2f}, {upper_bound:.2f}]")
    
    return outliers_mask, outliers_indices

# Trova outliers su entrambi gli assi
outliers_x_mask, outliers_x_indices = trova_outliers(x_data, 'X')
outliers_y_mask, outliers_y_indices = trova_outliers(y_data, 'Y')

# Combina: un punto è outlier se lo è su X O su Y
outliers_combined_mask = outliers_x_mask | outliers_y_mask
outliers_combined_indices = [i for i in range(len(x_data)) if outliers_combined_mask.iloc[i]]

print(f"\n--- Outliers Complessivi ---")
print(f"Totale outliers: {len(outliers_combined_indices)}")
print(f"Indici: {outliers_combined_indices}")
print(f"\nPer rimuoverli, usa questi indici nel tuo codice.")

# Crea lo scatter plot
plt.figure(figsize=(10, 6))

# Punti normali in blu
normal_mask = ~outliers_combined_mask
plt.scatter(x_data[normal_mask], y_data[normal_mask], 
           alpha=0.6, s=50, c='blue', label='Dati normali')

# Outliers in rosso
plt.scatter(x_data[outliers_combined_mask], y_data[outliers_combined_mask], 
           alpha=0.8, s=80, c='red', marker='x', label='Outliers')

# Annota gli indici degli outliers
for idx in outliers_combined_indices:
    plt.annotate(f'{idx}', 
                xy=(x_data.iloc[idx], y_data.iloc[idx]),
                xytext=(5, 5), textcoords='offset points',
                fontsize=8, color='red')

plt.xlabel('unfolding F')
plt.ylabel('W tot integral')
plt.title('MTdata_6')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('plot.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
