```python
import json
import os

def merge_pulse_json_files(file_paths, output_file=None):
    merged_data = {}
    pulse_counter = 0
    
    for file_path in file_paths:
        print(f"Processing {file_path}...")
        
        with open(file_path, 'r') as f:
            data = json.load(f)
        
        # Rinumera i pulse per evitare sovrapposizioni
        for key, value in data.items():
            if key.startswith("Pulse_Number_"):
                new_key = f"Pulse_Number_{pulse_counter}"
                merged_data[new_key] = value
                pulse_counter += 1
            else:
                merged_data[key] = value
    
    print(f"Total pulses merged: {pulse_counter}")
    
    if output_file:
        with open(output_file, 'w') as f:
            json.dump(merged_data, f, indent=2)
    
    return merged_data

# Uso:
json_files = [ 'MT3Data_JSON 5.Json', 'MT3Data_JSON 6.Json']
merged = merge_pulse_json_files(json_files, 'MT3Data_56.json')
```
